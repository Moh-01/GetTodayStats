## Importing libraries

```python
import re 
import requests
import pandas as pd
from datetime import datetime

#Scraping
from time import sleep
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.common.exceptions import NoSuchElementException

#Ignore unwanted warnings
import warnings
warnings.filterwarnings('ignore')
```

<br>

----

<br>

## Read webpage

```python
# Create a browser obj *This driver is for Chrome Version 91*
# If you have an older version or newer, download it and replace the path
driver = webdriver.Chrome(executable_path='chromedriver/chromedriver')

# Request url
driver.get('https://www.moh.gov.sa/Pages/Default.aspx')

# Delay 3s
sleep(30)

# Download page
html = driver.page_source
```

```python
# Get the html of the page
soup = BeautifulSoup(html,'html.parser')
```

```python
# Explore the webpage
soup
```

<br>

----

<br>

## Exploring tags

```python
# Data container
data = soup.find_all('div', attrs={'class':'col-sm-4 mb-4'})
data
```

<br>

----

<br>

## Collecting the data

### Total Examinations

```python
Examinations = int(data[5].find('div', attrs={'class':'counter red'}).string.replace(',', ''))
Examinations
```

### Total Vaccinations


```python
Vaccinations = int(data[0].find('div', attrs={'class':'counter red'}).string.replace(',', ''))
Vaccinations
```

### Total Active cases


```python
Active_cases = int(data[4].find('div', attrs={'class':'counter red'}).string.replace(',', ''))
Active_cases
```

### Total confirmed cases

```python
Total_conf_cases = int(re.sub(r"\D", "", data[1].find('h6', attrs={'class':'funfact-title'}).string))
Total_conf_cases
```

#### New cases

```python
New_cases = int(data[1].find('div', attrs={'class':'counter red'}).string.replace(',', ''))
New_cases
```

### Total recoveries

```python
Total_recoveries = int(re.sub(r"\D", "",data[2].find('h6', attrs={'class':'funfact-title'}).string))
Total_recoveries
```

#### New recoveries


```python
New_recoveries = int(data[2].find('div', attrs={'class':'counter red'}).string.replace(',', ''))
New_recoveries
```

### Total deaths

```python
Total_deaths = int(re.sub(r"\D", "",data[3].find('h6', attrs={'class':'funfact-title'}).string))
Total_deaths
```

#### New deaths


```python
New_deaths = int(data[3].find('div', attrs={'class':'counter red'}).string.replace(',', ''))
New_deaths
```

<br>

<br>

## Grouping the data

```python
COVID_moh = pd.DataFrame([{'Total_Examinations':    Examinations,
                           'Total_Vaccinations':    Vaccinations,
                           'New cases':             New_cases,
                           'Total_Active_cases':    Active_cases,
                           'Total_confirmed_cases': Total_conf_cases,
                           'New_recoveries':        New_recoveries,
                           'Total_recoveries':      Total_recoveries,
                           'New_deaths':            New_deaths,
                           'Total_deaths':          Total_deaths
                          }],
                         index = pd.to_datetime([datetime.now().strftime("%d/%m/%Y")]) # Set index to today's date
                        )
```
